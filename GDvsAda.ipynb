```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def f(w):
    return 0.5*w**2

def gradf(w):
    return w

# def eta_k(w):
#     return (3.0/4)/abs(gradf(w))
```

```python
iternum = 100
eta = 3.0/4
w_GD = 1
w_Ada = 1
F_GD = np.zeros(iternum)
F_Ada = np.zeros(iternum)
sum_grad_sq_ada = 0
for i in range(iternum):
    F_GD[i] = f(w_GD)
    F_Ada[i] = f(w_Ada)
    sum_grad_sq_ada += gradf(w_Ada)**2
    eta_i = (3.0/4)/math.sqrt(sum_grad_sq_ada)
    w_GD -= eta*gradf(w_GD)
    w_Ada -= eta_i*gradf(w_Ada)
```

```python
x = np.linspace(1,iternum, iternum, endpoint=True)
plt.plot(F_GD, label = 'GD')
plt.plot(F_Ada, label = 'Adaptive GD')
plt.yscale('log')
plt.xlabel('iterations')
plt.ylabel('log(F)')
plt.legend(loc='upper right')
plt.show()
```
